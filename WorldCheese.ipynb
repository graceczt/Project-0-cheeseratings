```python
#create a new column : to remove numerics on first 4 on 'category' column

worldcheesedf['cheese_category'] = worldcheesedf['category'].astype(str).str[6:]
```

```python
#delete category column
#data cleaning



worldcheesedf.drop(['category'], axis =1, inplace =True) 
```

```python
#drop county column
# as too many null values , only affects small amount of dataset

worldcheesedf.drop(columns = ['county'],inplace = True)


```

```python
#create 2nd df consisting of company verses rating COUNT, group by company,sort by highest rating only
worldcheesedf_count = worldcheesedf.groupby('country').count().sort_values(by=['rating'], ascending=False)['rating']

#create new table base on above
second_data = pd.DataFrame(data =worldcheesedf_count)

#to name columns 'company' and 'rating'
second_data = second_data.rename(columns={"rating": "count_of_awards"})
#to change index
second_data = second_data.reset_index()
#callable
second_data 




```

```python
#scatter plot

plt.scatter(x=second_data['country'], y=second_data['count_of_awards'])
plt.xticks(rotation=90)
plt.yticks(np.arange(min(second_data['count_of_awards']),max(second_data['count_of_awards']),20))
plt.title('country_VS_cheese_count_awards')
plt.xlabel('country')
plt.ylabel('cheese_count_awards')

plt.show()
```

```python
#create 3rd df consisting of country verses medals won, count by company
third_df = worldcheesedf.groupby(['rating','country']).count()['company']

#create new table base on above
medal_company = pd.DataFrame()

#to name columns 'company' and 'rating'
medal_company['Bronze medal'] = third_df['BRONZE']
medal_company['Silver medal'] = third_df['SILVER']
medal_company['Gold medal'] = third_df['GOLD']
medal_company['Super Gold medal'] = third_df['SUPER GOLD']
#medal_company = medal_company.reset_index()
medal_company
#to change index

#callable
```
